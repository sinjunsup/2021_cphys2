# Diabetes (clustering)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
diabetes=datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
X_p=diabetes.data[:,2] # 혈압
X_d = diabetes.target # 당뇨 진행 정도
X = [X_p,X_d]
plt.plot(X_p,X_d,'.')
```

# 1. sklearn을 이용하지 않고 (자신의 코드로)

```python
plt.scatter(X[0],X[1])
plt.plot([-0.02],[50],'r*')
plt.plot([0.05],[200],'b*')
plt.plot([0.1],[270],'k*')
```

```python
temp=[]
for (i,j) in zip(X[0],X[1]):
    r1=(i-(-0.02))**2 + (j-50)**2
    r2=(i-(0.05))**2 + (j-200)**2
    r3=(i-0.1)**2 + (j-270)**2
    m=min(r1,r2,r3)
    temp.append([r1,r2,r3].index(m))
```

```python
x1=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==0)])
x2=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==1)])
x3=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==2)])
r1=np.mean(x1[0]),np.mean(x1[1])
r2=np.mean(x2[0]),np.mean(x2[1])
r3=np.mean(x3[0]),np.mean(x3[1])
```

```python
plt.scatter(X[0],X[1],c=temp)
plt.plot(r1[0],r1[1],'r*')
plt.plot(r2[0],r2[1],'b*')
plt.plot(r3[0],r3[1],'k*')
```

# 2. sklearn을 이용해서

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
data = np.transpose(X)
kmeans.fit(data)
```

```python
plt.scatter(X[0],X[1],c=kmeans.labels_)
```

# 3. cluster 수를 결정하는 좋은 방법을 고안하여 적용해보아라.

위험등급 나누는 단계인 관심,주의,경고,위험,심각 총 5단계로 수를 나눔

```python
plt.scatter(X[0],X[1])
plt.plot([-0.02],[50],'g*')
plt.plot([-0.02],[100],'k*')
plt.plot([0.02],[150],'y*')
plt.plot([0.05],[200],'r*')
plt.plot([0.1],[270],'b*')
```

```python
temp=[]
for (i,j) in zip(X[0],X[1]):
    r1=(i-(-0.02))**2 + (j-50)**2
    r2=(i-(-0.02))**2 + (j-100)**2
    r3=(i-0.02)**2 + (j-150)**2
    r4=(i-0.05)**2 + (j-200)**2
    r5=(i-0.1)**2 + (j-270)**2
    m=min(r1,r2,r3,r4,r5)
    temp.append([r1,r2,r3,r4,r5].index(m))
```

```python
x1=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==0)])
x2=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==1)])
x3=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==2)])
x4=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==3)])
x5=np.transpose([[i,j] for (t,i,j) in zip(temp,X[0],X[1]) if(t==4)])
r1=np.mean(x1[0]),np.mean(x1[1])
r2=np.mean(x2[0]),np.mean(x2[1])
r3=np.mean(x3[0]),np.mean(x3[1])
r4=np.mean(x4[0]),np.mean(x4[1])
r5=np.mean(x5[0]),np.mean(x5[1])
```

```python
plt.scatter(X[0],X[1],c=temp)
plt.plot(r1[0],r1[1],'g*')
plt.plot(r2[0],r2[1],'k*')
plt.plot(r3[0],r3[1],'y*')
plt.plot(r4[0],r4[1],'r*')
plt.plot(r5[0],r5[1],'b*')
```
